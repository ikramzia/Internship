Q1. Write a python program which searches all the product under a particular product from www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.

```python
#importing Selenium, pandas, selenium webdriver, warnings, and time
import selenium
import pandas as pd
from bs4 import BeautifulSoup
import time

#importing selenium webdriver
from selenium import webdriver

#importing warnings
import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

#Importing Exceptions
from selenium.common.exceptions import NoSuchElementException as NSEE
from selenium.common.exceptions import StaleElementReferenceException as SERE
from selenium.common.exceptions import ElementNotInteractableException as ENIE
from selenium.common.exceptions import TimeoutException as TE
from selenium.common.exceptions import InvalidSelectorException as ISE
```

```python
amazon_driver = webdriver.Chrome() #opening automated chrome browser using selenium driver
amazon_driver.maximize_window() #maximize window
amazon_driver.get('https://www.amazon.in') #opening the webpage www.amazon.com on automated chrome browser
time.sleep(2) #asking driver to wait for 2 seconds

search_product = input("Enter any product to be searched on Amazon!\n") #taking user input to search for on amazon

#searching for required product by the user and exception handling
try:
    amazon_product = amazon_driver.find_element(By.XPATH,'//div[@class="nav-search-field "]/input')
    amazon_product.send_keys(search_product)
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)

try:
    amazon_search = amazon_driver.find_element(By.XPATH,'//span[@class="nav-search-submit-text nav-sprite nav-progressive-attribute"]/input')
    amazon_search.click()
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
    
try:
    search_result = amazon_driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-small a-spacing-top-small"]')
    print('Got %s'%search_result.text)
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
```

Q4) Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page.

Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”.

Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
#importing Selenium, pandas, selenium webdriver, warnings, and time
import selenium
import pandas as pd
from bs4 import BeautifulSoup
import time

#importing selenium webdriver
from selenium import webdriver

#importing warnings
import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

#Importing Exceptions
from selenium.common.exceptions import NoSuchElementException as NSEE
from selenium.common.exceptions import StaleElementReferenceException as SERE
from selenium.common.exceptions import ElementNotInteractableException as ENIE
from selenium.common.exceptions import TimeoutException as TE
from selenium.common.exceptions import InvalidSelectorException as ISE
```

```python
flipkart_driver = webdriver.Chrome() #opening automated chrome browser using selenium driver
flipkart_driver.maximize_window() #maximize window
flipkart_driver.get('https://www.flipkart.com') #opening the webpage www.flipkart.com on automated chrome browser
time.sleep(2) #asking driver to wait for 2 seconds

search_product = 'Oneplus Nord'

#searching for smartphone and exception handling
try:
    flipkart_product = flipkart_driver.find_element(By.XPATH,'//input[@class="Pke_EE"]')
    flipkart_product.send_keys(search_product)
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)

try:
    flipkart_search = flipkart_driver.find_element(By.XPATH,'//button[@class="_2iLD__"]')
    flipkart_search.click()
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
    
try:
    search_result = flipkart_driver.find_element(By.XPATH,'//span[@class="_10Ermr"]')
    print(search_result.text.replace("Showing", "Got"))
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
```

```python
#fetching smartphone detail from each smartphone url
flipkart_phone_detail_url = []
start = 0
end = 1

for flipkart_page in range(start, end):
    try:
        flipkart_url = flipkart_driver.find_elements(By.XPATH, '//a[@class="_1fQZEK"]')
        for i in flipkart_url:
            flipkart_phone_detail_url.append(i.get_attribute('href'))
    except NSEE:
        print("Exception Raised: \n", NSEE)
    except SERE:
        print("Exception Raised: \n", SERE)
    except ENIE:
        print("Exception Raised: \n", ENIE)
    except TE:
        print("Exception Raised: \n", TE)
    except ISE:
        print("Exception Raised: \n", ISE)
```
